# Análisis exploratorio de datos con enfasis en Mathematics Stack Exchange

## 1. Introducción

### Abstract

##### De acuerdo con su sitio web, Mathematics Stack Exchange es un sitio/foro de preguntas y respuestas para personas que estudian matemáticas en cualquier nivel y profesionales en campos relacionados.
##### Es una excelente plataforma (que por cierto es gratuita) para las personas con problemas acerca de algún área de las matemáticas en específico, ya que gente que diversas partes del mundo pueden visualizar e intentar responder tu pregunta mientras tú haces lo mismo.
##### El objetivo principal del análisis es identificar patrones y tendencias en las preguntas, vistas y respuestas.
##### Link al sitio: https://math.stackexchange.com/

## 3. Conclusiones

##### La información dada refleja ciertas como lo son:
##### Se trata de una plataforma la cual no es muy conocida por toda la comunidad, por lo que era de esperarse pocas respuestas a las preguntas (Cerca del 7%), por lo que no es muy recomenada en situaciones donde surga alguna duda o pregunta que necesite ser respondida como urgencia, si no más bien como un foro para compartir conocimientos por hobbie, o tomar preguntas que surjan en el sitio como retos personales para resolver en tiempo libre para aquellos amantes de las matemáticas.
##### Además (de forma lógica) las preguntas con pocas vistas tienen pocas o nulas respuestas, mientras que las preguntas con una cantidad considerable de vistas tienen más respuestas.
